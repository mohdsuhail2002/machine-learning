## Task4:

Create histogram with Kernel Density Estimation (KDE) plots for both the 'age' and 'player_rating' columns from the dataset. While constructing these plots, ensure to visually represent the standard deviation intervals on the graph.


- **'age'**

```python
col = 'age'
mean = df[col].mean()
std = df[col].std()

plt.figure(figsize=(10, 5))
sns.histplot(df[col], bins=10, kde=True, color='skyblue')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='+1 SD')
plt.axvline(mean - std, color='green', linestyle='--', label='-1 SD')
plt.axvline(mean + 2*std, color='orange', linestyle='--', label='+2 SD')
plt.axvline(mean - 2*std, color='orange', linestyle='--', label='-2 SD')
plt.axvline(mean + 3*std, color='purple', linestyle='--', label='+3 SD')
plt.axvline(mean - 3*std, color='purple', linestyle='--', label='-3 SD')
plt.title(f'{col.capitalize()} Distribution with KDE & 3 SD')
plt.legend()
plt.show()
```

- **'player_rating'**

```python
col = 'player_rating'
mean = df[col].mean()
std = df[col].std()


plt.figure(figsize=(10, 5))
sns.histplot(df[col], bins=10, kde=True, color='lightcoral')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='+1 SD')
plt.axvline(mean - std, color='green', linestyle='--', label='-1 SD')
plt.axvline(mean + 2*std, color='orange', linestyle='--', label='+2 SD')
plt.axvline(mean - 2*std, color='orange', linestyle='--', label='-2 SD')
plt.axvline(mean + 3*std, color='purple', linestyle='--', label='+3 SD')
plt.axvline(mean - 3*std, color='purple', linestyle='--', label='-3 SD')
plt.title(f'{col.replace("_", " ").capitalize()} Distribution with KDE & 3 SD')
plt.legend()
plt.show()
```

## Task5

- Calculate the lower and upper boundaries for the 'Age' and 'Player_Rating' columns. 
- For each column, identify values that fall outside the calculated lower and upper boundaries. These values are considered outliers.

- **Display age outliers**

```python
def get_bounds_std(data, num_std=3):
    mean = data.mean()
    std = data.std()
    lower = mean - num_std * std
    upper = mean + num_std * std
    return lower, upper
lower_age, upper_age = get_bounds_std(df['age'], num_std=3)
lower_rating, upper_rating = get_bounds_std(df['player_rating'], num_std=3)
```

- **Display player_rating outliers**

```python
df_outlier=df[(df["player_rating"]<lower_rating) | (df["player_rating"]>upper_rating)]
df_outlier
```

## Task6

-  Remove outliers from both the 'age' and 'player_rating' columns and then visualize the adjusted data using histogram plots.


```python
df_new1=df[(df["player_rating"]>lower_rating) & (df["player_rating"]<upper_rating)]
df_new1
```

```python
df_outlier1=df[(df["age"]<lower_age) | (df["age"]>upper_age)]
df_outlier1
```

```python
df_final=df_new1[(df_new1["age"]>lower_age) & (df_new1["age"]<upper_age)]
df_final
```

- **plot the histograms**

```python
col = 'age'
mean = df_final[col].mean()
std = df_final[col].std()

plt.figure(figsize=(10, 5))
sns.histplot(df_final[col], bins=10, kde=True, color='skyblue')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='+1 SD')
plt.axvline(mean - std, color='green', linestyle='--', label='-1 SD')
plt.axvline(mean + 2*std, color='orange', linestyle='--', label='+2 SD')
plt.axvline(mean - 2*std, color='orange', linestyle='--', label='-2 SD')
plt.axvline(mean + 3*std, color='purple', linestyle='--', label='+3 SD')
plt.axvline(mean - 3*std, color='purple', linestyle='--', label='-3 SD')
plt.title(f'{col.capitalize()} (After Outlier Removal)')
plt.legend()
plt.show()
```
